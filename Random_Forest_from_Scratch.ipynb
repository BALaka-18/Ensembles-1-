```
class RandomForestClassifier():
  def __init__(self,n_estimators=100,depth=None,criterion='gini',max_features='auto',bootstrap=True,max_samples=None):
    self.n_estimators = n_estimators
    self.depth = depth
    self.criterion = criterion
    self.max_features = max_features
    self.bootstrap = bootstrap
    self.max_samples = max_samples
  def fit(self,trainx,trainy):
    global trees
    trainy_final,trees = [],[]
    for _ in n_estimators:
      if self.bootstrap == True:
        if self.max_samples != None:
          new_trainx = trainx.sample(n=self.max_samples,replace=True)
          vals = list(trainx.values)
          i = 0
          for v in list(new_trainx.values):
            while i < len(vals):
              if set(list(v)) == set(list(vals[i])):
                trainy_final.append((list(trainy))[i])
                break
              else:
                i = i + 1
          if self.max_features == 'auto' or self.max_features == 'sqrt':
            trainx_final = new_trainx.sample(n=np.sqrt(new_trainx.shape[1]),replace=True,axis=1)
          elif self.max_features == 'log2':
            trainx_final = new_trainx.sample(n=np.log2(new_trainx.shape[1]),replace=True,axis=1)
          elif type(self.max_features) == 'int':
            trainx_final = new_trainx.sample(n=self.max_features,replace=True,axis=1)
          else:
            trainx_final = new_trainx
        else:
          new_trainx = trainx
          if self.max_features == 'auto' or self.max_features == 'sqrt':
            trainx_final = new_trainx.sample(n=np.sqrt(new_trainx.shape[1]),replace=True,axis=1)
          elif self.max_features == 'log2':
            trainx_final = new_trainx.sample(n=np.log2(new_trainx.shape[1]),replace=True,axis=1)
          elif type(self.max_features) == 'int':
            trainx_final = new_trainx.sample(n=self.max_features,replace=True,axis=1)
          else:
            trainx_final = new_trainx
      else:
        trainx_final = trainx
        trainy_final = trainy
      clf = DecisionTreeClassifier(max_depth=self.depth,criterion=self.criterion)
      trees.append(clf.fit(trainx_final,trainy_final))
  def predict(self,testx):
    global trees
    agg,freq = [],{}
    for tree in trees:
      agg.append(tree.predict(testx))
    for item in agg:
      if item in freq.keys():
        freq[item] += 1
      else:
        freq[item] = 1
    labels,count = list(freq.items()),list(freq.values())
    return labels[max(count).index]
```
